```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Get the path to the folder containing the X-ray images
path_to_folder = "C:\\mini project\\dataset\\train\\4"

# Get a list of all the files in the folder
files = os.listdir("C:\\mini project\\dataset\\train\\4")

# Create a list to store the X-ray images
x_rays = []

# Loop through the files in the folder
for file in files:

    # Check if the file is an image
    if file.endswith('.jpg') or file.endswith('.png'):

        # Read the image
        image = cv2.imread(os.path.join("C:\\mini project\\dataset\\train\\4", file))

        # Resize the image to a consistent size
        image = cv2.resize(image, (224, 224))

        # Convert the image to a NumPy array
        image = np.array(image)

        # Add the image to the list of X-ray images
        x_rays.append(image)

# Convert the list of X-ray images to a NumPy array
x_rays = np.array(x_rays)

# Display the first X-ray image
plt.imshow(x_rays[0])
plt.show()
```
